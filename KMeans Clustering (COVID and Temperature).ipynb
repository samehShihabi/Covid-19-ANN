```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df=pd.read_csv('5factors and new cases.csv')
```

```python
df
```

```python
df[['TemperatureAVG','COVIDCASES']]
```

```python
plt.scatter(df['TemperatureAVG'],df['COVIDCASES'])
plt.ylabel('COVIDCASES')
plt.xlabel('Average Temperature (C)')
```

```python
km=KMeans(n_clusters=2)
km
```

```python
y_predicted=km.fit_predict(df[['TemperatureAVG','COVIDCASES']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1= df[df.cluster==0]
df2= df[df.cluster==1]

plt.scatter(df1.TemperatureAVG,df1['COVIDCASES'], color='black')
plt.scatter(df2.TemperatureAVG,df1['COVIDCASES'], color='red')

```

```python
scaler= MinMaxScaler()
scaler.fit(df[['COVIDCASES']])
df['COVIDCASES'] == scaler.transform(df['COVIDCASES'])

scaler.fit(df.TemperatureAVG)
df.TemperatureAVG= scaler.transform(df.TemperatureAVG)
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['TemperatureAVG','COVIDCASES']])
y_predicted
```

```python
df['cluster']= y_predicted
df.drop('cluster', axis='columns', inplace=True)
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.TemperatureAVG, df1['COVIDCASES'], color='blue')
plt.scatter(df2.TemperatureAVG, df2['COVIDCASES'], color='green')
plt.scatter(df3.TemperatureAVG, df3['COVIDCASES'], color='black')
plt.legend()
```
