```python
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import matplotlib.pyplot as plt
```

```python
JordanCases=JordanDF[["new_cases"]]
UAECases=UAEDF[["new_cases"]]
UAECases.shape
```

```python
scalarJordan=MinMaxScaler()
scalarJordan.fit(JordanCases)
normalizedJordanCases=scalarJordan.transform(JordanCases)
#print(normalizedJordanCases)
#normalizedJordanCase
scalarUAE=MinMaxScaler()
scalarUAE.fit(UAECases)
normalizedUAECases=scalarUAE.transform(UAECases)
print(len(normalizedUAECases))
```

```python
n_input=5
n_output=1
generatorJordan=TimeseriesGenerator(normalizedJordanCases,normalizedJordanCases,length=n_input,batch_size=1)
#print(generatorJordan[0])
#print(generatorJordan[1])
generatorUAE=TimeseriesGenerator(normalizedUAECases,normalizedUAECases,length=n_input,batch_size=1)
#print(generatorUAE[0])
#print(generatorUAE[1])
print(len(generatorUAE))
```

```python
dataJordan=[]
targetJordan=[]
for i in range(len(generatorJordan)):
    data1,target1=generatorJordan[i]
    dataJordan.append(data1[0])
    xx=float(target1[0])
    targetJordan.append(xx)
dataJordan=np.array(dataJordan)
targetJordan=np.array(targetJordan)
x_train_Jordan,x_test_Jordan,y_train_Jordan,y_test_Jordan=train_test_split(dataJordan,targetJordan,test_size=0.2,random_state=4)
print(dataJordan.size)
print(x_test_Jordan.size)
```

```python
dataUAE=[]
targetUAE=[]
for i in range(len(generatorUAE)):
    data2,target2=generatorUAE[i]
    dataUAE.append(data2[0])
    xx=float(target2[0])
    targetUAE.append(xx)
dataUAE=np.array(dataUAE)
targetUAE=np.array(targetUAE)
x_train_UAE,x_test_UAE,y_train_UAE,y_test_UAE=train_test_split(dataUAE,targetUAE,test_size=0.2,random_state=4)
type(y_test_UAE)
print(x_test_UAE.size)
print(dataUAE.size)
```

```python
model=Sequential()
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences=False))
#model.add(LSTM((1),return_sequences=True))
#model.add(LSTM((1),return_sequences=False))
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
history=model.fit(x_train_Jordan,y_train_Jordan,epochs=400,validation_data=(x_test_Jordan,y_test_Jordan))
```

```python
plt.plot(history.history['loss'])
plt.show()
```

```python
results_Jordan=model.predict(x_test_Jordan)
print(len(results_Jordan))

plt.scatter(range(len(y_test_Jordan)),y_test_Jordan,c='r')
plt.scatter(range(len(results_Jordan)),results_Jordan,c='g')
plt.show()
```

```python
resultsNotScaledJordan=scalarJordan.inverse_transform(results_Jordan)
y_test2_Jordan=y_test_Jordan.reshape(-1,1)
targetNotScaledJordan=scalarJordan.inverse_transform(y_test2_Jordan)
type(results_Jordan)
type(y_test_Jordan)
plt.scatter(range(len(targetNotScaledJordan)),targetNotScaledJordan,c='r')
plt.scatter(range(len(resultsNotScaledJordan)),resultsNotScaledJordan,c='g')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(resultsNotScaledJordan,targetNotScaledJordan))
print(rmse)
```
